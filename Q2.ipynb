```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

# Load the brain image in grayscale (ensure the correct path to the image)
image = cv.imread('a1images/a1images/brain_proton_density_slice.png', cv.IMREAD_GRAYSCALE)

# Define the intensity transformation function to highlight gray matter
def intensity_transformation(value):
    if 190 <= value <= 255:  # Target gray matter intensity range
        return int(1.5 * value)  # Amplify gray matter intensity by 1.5x
    else:
        return value  

# Apply the transformation to the image by iterating through each pixel
transformed_image = np.zeros_like(image)
for i in range(image.shape[0]):
    for j in range(image.shape[1]):
        transformed_image[i, j] = intensity_transformation(image[i, j])

# Display the original and transformed images side by side
plt.subplot(121), plt.imshow(image, cmap='gray'), plt.title('Original Image')
plt.subplot(122), plt.imshow(transformed_image, cmap='gray'), plt.title('Transformed Image (Gray Matter Highlighted)')
plt.show()

# Plot the transformation function to visualize the changes in intensity
x = np.arange(0, 256)  # Input intensity values
y = [intensity_transformation(i) for i in x]  # Transformed intensity values
plt.plot(x, y)
plt.title('Gray Matter Intensity Transformation')
plt.xlabel('Input Intensity')
plt.ylabel('Output Intensity')
plt.show()
```

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

# Load the brain image (ensure the correct path)
image = cv.imread('a1images/a1images/brain_proton_density_slice.png', cv.IMREAD_GRAYSCALE)

# Define the intensity transformation function to accentuate gray matter
def intensity_transformation(value):
    if 190 <= value <= 255:  # Gray matter intensities
        return min(255, int(2 * value))  # Enhance gray matter (boost intensities)
    else:
        return value  

# Apply the transformation to the image
transformed_image = np.zeros_like(image)
for i in range(image.shape[0]):
    for j in range(image.shape[1]):
        transformed_image[i, j] = intensity_transformation(image[i, j])

# Display the transformed image
plt.subplot(121), plt.imshow(image, cmap='gray'), plt.title('Original Image')
plt.subplot(122), plt.imshow(transformed_image, cmap='gray'), plt.title('Transformed Image (White Matter Accentuated)')
plt.show()

# Plot the transformation function
x = np.arange(0, 256)
y = [intensity_transformation(i) for i in x]
plt.plot(x, y)
plt.title('White Matter Intensity Transformation')
plt.xlabel('Input Intensity')
plt.ylabel('Output Intensity')
plt.show()
```
